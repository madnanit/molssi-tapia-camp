```python
import os
import numpy
```

```python
distance_file = os.path.join('distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',', dtype='unicode')
headers = distances[0,:]
data = distances[1:,:]
data = data.astype(numpy.float)
                                # similar to yesterday. we are reading the file and separaating it into header and data
```

```python
print(headers)
```

```python
print(data)
```

```python
import matplotlib.pyplot as plt
                                # we're importing a module called "matplotlib.pyplot" but
                                #since the name is too long we're giving it a short name "plt"
```

```python
plt.plot(data[:,1])
                                # we're plotting the 1st column

plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
                                # adding labels to x and y axis
```

```python
# alternative way to label axes

sample = headers[1]
                                # using header file to label axis

plt.plot(data[:,1], label=sample)
                                # we're plotting the 1st column and labeling it by "sample"

plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
                                # adding labels to x and y axis
    
plt.legend()
                                # adding legend
    
plt.savefig(F'{sample}.png', dpi=300)
                                # saving the plot as "sample.png" and specifying the dpi=300
```

```python
sample = headers[1]
sample2 = headers[2]
                                # using header file to label axis

plt.plot(data[:,1], label=sample)
plt.plot(data[:,2], label=sample2)
                                # we're plotting the 1st and 2nd column and labeling it by "sample" and "sample2"

plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
                                # adding labels to x and y axis
    
plt.legend()
                                # adding legend
    
plt.savefig('two_samples.png', dpi=300)
                                # saving the plot as "sample.png" and specifying the dpi=300
```

```python
num_columns = len(headers)

print(headers)
print(F'The number of columns are= {num_columns}')

for col in range(1, num_columns):
    
    print(F'The number of columns are= {col}')
    plt.plot(data[:,col], label=headers[col])

plt.legend() 
plt.xlabel('Simulation Frame')
plt.ylabel('Distances (angstrom)')
plt.savefig('all_samples.png')
```

```python
num_columns = len(headers)

print(headers)
print(F'The number of columns are= {num_columns}')


for col in range(1, num_columns):
    
    plt.figure()
                                                    # it creates a new figure for each column
        
    plt.plot(data[:,col], label=headers[col])
    plt.legend() 
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distances (angstrom)')
    plt.savefig(F'{headers[col]}.png')
```
